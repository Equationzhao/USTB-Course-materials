### 汇编语言

汇编语言是第二代计算机语言，指一些容易理解和记忆的字母、单词来代替一个特定的指令。而指令则是用于控制计算机进行各种操作的01序列。

当C语言被编译成可执行程序时，经过了预编译、编译、汇编、链接四个过程，其中编译过程就是将C语言代码转换为汇编语言代码，而汇编过程是将汇编代码转换成01序列，最后将01序列进行链接，就生成了操作系统上可以运行的二进制可执行程序。而拆弹实验的基本目的，就是通过可执行程序反推出汇编代码，再通过阅读汇编代码理解程序运行过程。其根本目的是通过理解汇编语言，将高级语言与机器语言的知识点串联起来，同时，学会阅读汇编语言，更方便理解高级语言的底层含义，提升学生debug能力。

汇编语言主要分为Intel风格和AT&T风格，我们使用objdump命令产生的汇编代码是AT&T风格，而使用编译器生成的汇编代码是Intel风格。它们之间最大的差别是在INTEL语法中，第一个表示目的操作数，第二个表示源操作数，赋值方向从右向左。AT&T语法第一个为源操作数，第二个为目的操作数，方向从左到右。例如：

> (Intel) MOV EAX, EBX //将EBX寄存器的内容移入EAX寄存器中

> (AT&T) movl %ebx,%eax//将EBX寄存器的内容移入EAX寄存器中

如果你在学习汇编的过程中发现各版本的说法不一致，请检查其风格。


### 寄存器

注意在上述代码中我们使用了eax和ebx两个寄存器，这是在早期8086架构CPU的基础上衍生而来的。

早期的8086CPU总共有14个寄存器，它们各有用处：

AX、BX、CX、DX、SP、BP、SI、DI、IP、FLAG、CS、DS、SS、ES

它们在时代的进化过程中经过了许多变化，例如如今32位x86如下表示：

EAX、EBX、ECX、EDX、
ESP、EBP、
ESI、EDI、EIP、
ES、CS、SS、DS、FS、GS


其中，EAX常用于暂存变量、累加，以及作为函数返回值。EBX常用于储存内存基地址(base pointer)，ESP和EBP为一对指针寄存器，用于在函数调用时存储栈地址(stack pointer)和栈帧基地址(frame pointer)。EIP为指令指针寄存器，用于存储下一条指令的地址。虽然目前已经很少见到32位处理器了，但是由于本次拆弹实验使用32位汇编，因此我们不对64位汇编进行过多介绍。


### 栈帧

在函数调用(call指令)时，首先将下一条指令的地址(eip的值)压入栈，然后将调用的函数地址更新至eip中，接着将原栈帧的基地址(ebp的值)压入栈，然后将现在的esp作为新的ebp，此时进入新的函数。

在函数返回时，eax中保存着函数的返回值，接着执行ret命令，基本为call命令的逆过程，首先将esp恢复为ebp（释放函数中使用的内存空间），接着取出栈顶保存的值放回ebp中，再将下一个栈顶的值放入eip中，就完成了函数的返回。

注1：当压入栈时，esp会自动移动至新的栈顶（指向仍未存放数据的空间）。

栈是汇编语言的一个重要概念，它用于实现函数的递归调用。操作系统在启动可执行程序时会为其分配内存空间，其中包括栈。栈有固定的大小，在栈帧中储存着许多的变量，以ebp加上偏移量进行访问。栈帧由内存高地址向低地址增长，以ebp近似作为两个函数栈内存空间的划分。

例如，我们看看phase 0的汇编代码：

```armasm
push  %ebp
mov   %esp,%ebp
sub   $0x8,%esp

sub   $0x8,%esp
push  $0x804a1e0
pushl 0x8(%ebp)
call  8049c4b <strings_not_equal>

test  %eax,%eax
je    8049478 <phase_0+0x29>
call  8049eb3 <explode_bomb>

mov   $0x0,%eax
jmp   804947d <phase_0+0x2e>
mov   $0x1,%eax
leave
ret
```

其中，就有`pushl 0x8(%ebp)`，它意味着我们将ebp的值加上0x8以后的内存地址，即从ebp往栈底（上个函数调用phase_0时传递的参数）走8个字节，取该内存地址中保存的值压入栈顶。而这8个字节代表的64位，正好是函数调用时存入的eip和ebp。而在该命令后紧接着一个call strings_not_equal，如果你尝试着在纸上画一下当前的栈帧，并模拟call指令的过程，你会发现在strings_not_equal函数中，刚刚压入栈的变量变成了0x8(%ebp)，即上个函数传递进来的参数。至于这个变量为何物，我们只能查看main函数中是如何调用phase_0函数或者是通过gdb查看内存了。

除了栈以外，操作系统还存在堆空间，用于动态申请内存空间，还有专门的内存空间用于存放可执行程序的代码段、数据段等。


### 基本命令

常用的命令简述如下：(imm代表立即数，mem代表内存中某个地址中的变量)

```armasm
lea  mem,%eax  将mem代表的元素的内存地址放入eax中
mov  %eax,%ebx 将eax的值放入ebx中
call 0x80490cb 调用内存地址0x80490cb处的函数
add  %eax,%ebx 理解成高级语言就是 ebx+=eax
sub  %eax,%ebx 理解成高级语言就是 ebx-=eax
jmp  imm       执行完毕后跳转到ins处继续执行程序
cmp  %eax,%ebx 计算eax和ebx的差值并判断结果，修改标志位（例如，结果为0，则ZF置1）
test %eax,%eax 计算eax和eax的与运算并判断结果，修改标志位
je   imm       若ZF=1则跳转
jne  imm       若ZF=0则跳转
jg   imm       若之前cmp大于则跳转（无符号数）
```

至于具体的规则，由于本实验不要求学生们自己手写汇编指令，因此我们不在此详细展开讲。各位只需要了解基本语法，再根据基本意思可以推断其语法，加上搜索引擎的使用，应该足以顺利完成本实验。

下面，我们补充一下指令中各种参数形式：

%eax 表示寄存器内储存的值

$0x0a 表示立即数0x0a

(%eax) 表示以寄存器内储存的值为地址，在内存中寻址得到的值

imm(%eax) 表示以寄存器内储存的值，加上ins偏移量后的地址，在内存中寻址得到的值

imm1(%eax,%ebx,imm2) 表示内存地址为eax+ebx*imm2+imm1的值

## 拆弹过程示例

---

我们将以phase_0为例进行拆弹过程的演示：

```armasm
0804944f <phase_0>:
 804944f:       55                      push   %ebp
 8049450:       89 e5                   mov    %esp,%ebp
 8049452:       83 ec 08                sub    $0x8,%esp
 8049455:       83 ec 08                sub    $0x8,%esp
 8049458:       68 e4 a1 04 08          push   $0x804a1e4
 804945d:       ff 75 08                pushl  0x8(%ebp)
 8049460:       e8 bd 07 00 00          call   8049c22 <strings_not_equal>
 8049465:       83 c4 10                add    $0x10,%esp
 8049468:       85 c0                   test   %eax,%eax
 804946a:       74 0c                   je     8049478 <phase_0+0x29>
 804946c:       e8 19 0a 00 00          call   8049e8a <explode_bomb>
 8049471:       b8 00 00 00 00          mov    $0x0,%eax
 8049476:       eb 05                   jmp    804947d <phase_0+0x2e>
 8049478:       b8 01 00 00 00          mov    $0x1,%eax
 804947d:       c9                      leave
 804947e:       c3                      ret
```

首先纵观整段代码，以`push %ebp`和`mov %esp,%ebp`为起始，以`leave`和`ret`为终止，之间是函数的内容。

`sub $0x8,%esp`是开辟内存空间而未赋初值，可以理解成C语言代码的`int a;`

`push $0x80a1e4`和`push 0x8(%ebp)`后紧接着一个`call 8049c22`，那么这两个被push进去的值很可能就是被调用函数的参数。

我们观察被调用的函数名为strings_not_equal，那么这两个参数应该是两个字符串，$0x80a1e4中是程序内存储的字符串字面值，而0x8(%ebp)是由main函数向phase_0函数传递的参数，想必就是同学们输入的字符串。在函数返回后，`test eax,eax`用来判断eax是否为0，若eax为0，则将标志位ZF置1，才满足je跳转的条件，否则将会执行explode_bomb函数。

根据上述分析，我们猜测，我们应该需要输入与0x804a1e4中保存的字符串相同的字符串，使得strings_not_equal函数返回false，即令eax为0，令je成功跳转避免爆炸。

因此，使用gdb查看该字符串：

```bash
(gdb) x/s 0x804a1e4
0x804a1e4: "Unix system programming is an old art."
```

我们将该字符串写入solution.txt的第一行，phase_0拆除完毕。
